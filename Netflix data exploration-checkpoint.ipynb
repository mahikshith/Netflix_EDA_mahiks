Handling the duplicate , NaN values and replacing those with "Data_Not_Available"

And for country column replaced the missing values with mode of all the countires

```python
# Want to know the number of  NaN values present in the data frame
df.isnull().sum()
```

```python
print(df.columns,len(df))
```

```python
# Percentage of NaN contribution from the above  :
# director has the highest NaN contribution about 30  percent

for each in df.columns:
  ans = (df[each].isnull().sum()/len(df)) *100 
  if ans >0:
    print(each,"---",ans)
```

```python
# We can see that all the missing values are genric in nature and doesn't really add value to the data frame
# So we can use "Data_Not_available" in the place of NaN.
df['country'] = df['country'].fillna(df['country'].mode()[0])
df['cast'].replace(np.nan,'Data_Not_available',inplace=True)
df['director'].replace(np.nan,'Data_Not_available',inplace=True)
```

```python
# Verifying the same:
df["country"].isnull().sum()
df["director"].isnull().sum()
```

```python
df.duplicated().sum()
# Total duplicate values are found to be zero
```

We can see that the "date added" data type is object.
Converting it into datetime data type

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df.info()
```

**Analysis**

```python
x=df["type"].value_counts()
print(x)
sns.barplot(x=x.index,y=x.values)
```

Obviously people like to spend 2 hours on a movie rather than 20 hours on a TV-show with more seasons. So production of movies is higher than TV-shows

```python
# Finding the relation between the type and Rating
plt.figure(figsize=(10,10))
sns.countplot(data=df,x='rating',hue='type')
plt.title('Type and Rating relation')
plt.show()
```

It could be inferred from the above that the most of the subscribers enjoy  TV-MA rated movies and TV shows. However, interesting thing is that Most people enjoy "R" rated movies than Tv shows of same rating.

```python
df['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,10))
plt.show()
```

TV-MA – TV Mature Audience Only 
Means high amount of content is target for mature audience who are above the age group 17 , which indicates that around 36 %  of the Netflix's audience / subscribers are above 17 or enjoy mature content more.

```python
plt.figure(figsize=(10,10))
sns.countplot(y="release_year", data=df, palette="Set2", order=df['release_year'].value_counts().index[0:15])
```

Netflix released / produced most of the content in the year 2018 rather than 2019 and 2020.
Which can be inferred as the GROWTH year for netflix in heavily competitive market.

```python
plt.figure(figsize=(10,10))
country_plot=df["country"].value_counts(ascending=False)
country_plot=country_plot.head(10)
ax=sns.barplot(x=country_plot.index,y=country_plot.values)
ax.tick_params(axis='x', rotation=90)
ax.set(xlabel='Country', ylabel='Number')
print(country_plot)
```

From the above plot, we can understand that the USA is very important market place for netflix. Hence Netflix's primary content is made USA market.

```python
genre=df["listed_in"].apply(lambda x: str(x).split(",")).tolist()
genre[2]
df_genre=pd.DataFrame(genre,index=df["title"])
df_genre=df_genre.stack()
df_genre=pd.DataFrame(df_genre.reset_index())

df_genre.rename(columns={0:"Genres"},inplace=True)
df_genre.drop(["level_1"],axis=1,inplace=True)
```

```python
df_genre=df_genre.merge(df,on="title")
df_genre.loc[df_genre["Genres"]==" Dramas","Genres"]="Dramas"
#corecting spelling
```

```python
a=df_genre.loc[df_genre["Genres"]!="Missing Value","Genres"].value_counts()
a=a.head(10)
print(a)
```

```python
plt.figure(figsize=(10,10))

a=df_genre.loc[df_genre["Genres"]!="Missing Value","Genres"].value_counts()
a=a.head(10)
axs=sns.barplot(x=a.index,y=a.values)
axs.tick_params(axis='x', rotation=90)
```

People like International movies followed by dramas

So, In the next fiscal year it is recommended to make more International movies and dub them in other regions while maintaining the count of other generes as well.

```python
plt.figure(figsize=(8,8))

df["day_of_week"]=pd.to_datetime(df["date_added"]).dt.day_name()

dfday=df["day_of_week"]
dfday=dfday.sort_values()
axs=sns.histplot(data=dfday,x=dfday.values)
axs.tick_params(axis='x', rotation=90)
axs.set(xlabel="Day of the Week",ylabel="Total Number")
```

Since most of the people enoy friday evenings with their families , it recommended to release new movies on friday
